**Aggregate functions**

```python
%%sql result <<
SELECT COUNT(*) AS total_employees, 
FROM Employee;
```

```python
%%sql result <<
SELECT SUM(Milliseconds)/60000 AS total_minutes,
AVG(Milliseconds)/1000 AS avg_seconds,
MIN(UnitPrice) AS cheapest_unit,
MAX(UnitPrice) AS priciest_unit
FROM Track;
```

```python
result
```

**GROUP BY**

```python
%%sql result <<
SELECT TrackId, Composer
FROM Track
GROUP BY Composer;
```

```python
result
```

**HAVING**

```python
%%sql result <<
SELECT SUM(UnitPrice) AS total_price, Composer
FROM Track
GROUP BY Composer
HAVING total_price > 10;
```
